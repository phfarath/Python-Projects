#### Calculando o coeficiente angular e linear

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
# Calculando coeficiente angular 
```

```python
b = np.mean(Y)- a*np.mean(X)
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow-y)
# Geramos uma regressão linear com a norma
```

### Motivo da reta

```python
plt.plot(datas, Moscow)
plt.plot(x,y)
# estimando valores
plt.plot(41.5, 41.5*a+b, '*r')
# estimando valores futuros 
plt.plot(100,100*a+b, '*r')
```
