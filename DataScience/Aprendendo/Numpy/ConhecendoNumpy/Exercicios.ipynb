### Após fazer a seleção de dados, importe a biblioteca matplotlib e crie um gráfico para a laranja e para a toranja do peso pelo diâmetro.

```python
import numpy as np
```

```python
diametro_laranja = np.loadtxt('citrus.csv', delimiter=',', usecols=(1), skiprows=1)
peso_laranja = np.loadtxt('citrus.csv', delimiter=',', usecols=(2), skiprows=1)

diametro_toranja = np.loadtxt('citrus.csv', delimiter=',', usecols=(1), skiprows=5000)
peso_toranja = np.loadtxt('citrus.csv', delimiter=',', usecols=(2), skiprows=5000)

import matplotlib.pyplot as plt
plt.plot(diametro_laranja, peso_laranja, label='Laranja')
plt.plot(diametro_toranja, peso_toranja, label='Toranja')
plt.xlabel('Diâmetro')
plt.ylabel('Peso')
plt.legend()
plt.show()
```

### Continuando com o projeto das laranjas/toranjas agora você deve calcular o coeficiente ângular e o linear para a reta da laranja e para a reta da toranja. Use a fórmula de mínimos quadrados para encontrar cada um.

```python
Y_lar = peso_laranja
X_lar = diametro_laranja
n_lar = np.size(peso_laranja)
```

```python
a_laranja = (n_lar*np.sum(X_lar*Y_lar) - np.sum(X_lar)*np.sum(Y_lar))/(n_lar*np.sum(X_lar**2) - np.sum(X_lar)**2)
a_laranja
```

```python
b_laranja = np.mean(Y_lar)- a_laranja*np.mean(X_lar)
```

```python
y_lar = a_laranja*X_lar + b_laranja
```

```python
regressao_laranja = np.linalg.norm(peso_laranja-y_lar)
plt.plot(diametro_laranja, peso_laranja, label='Laranja')
plt.plot(X_lar,y_lar, color='red')
```

```python
Y_tor = peso_toranja
X_tor = diametro_toranja
n_tor = np.size(peso_toranja)
```

```python
a_tor = (n_tor*np.sum(X_tor*Y_tor) - np.sum(X_tor)*np.sum(Y_tor))/(n_tor*np.sum(X_tor**2) - np.sum(X_tor)**2)
a_tor
```

```python
b_tor = np.mean(Y_tor)- a_tor*np.mean(X_tor)
```

```python
y_tor = a_tor*X_tor + b_tor
```

```python
regressao_toranja = np.linalg.norm(peso_toranja-y_tor)
plt.plot(diametro_toranja, peso_toranja, label='Toranja')
plt.plot(X_tor,y_tor, color='red')
```
