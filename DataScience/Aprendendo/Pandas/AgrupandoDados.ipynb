### UNINDO DADOS 

```python
censo_demo = pd.read_excel('dataset/POP2022_Municipios.xls', header=1, skipfooter=34)
censo_demo
```

```python
censo_demo.groupby('UF').sum(numeric_only=True)
```

```python
censo_demo['POPULAÇÃO'].astype(int)
```

```python
# filtrando os dados com .str.contains()
censo_demo[censo_demo['POPULAÇÃO'].str.contains('\(', na=False)]
```

```python
censo_demo = censo_demo.assign(populacao_sem_parenteses = censo_demo['POPULAÇÃO'].replace('\(\d{1,2}\)', '', regex = True),
                                             populacao = lambda x : x.loc[:, 'populacao_sem_parenteses'].replace('\.', '', regex = True))
```

```python
censo_demo[censo_demo['POPULAÇÃO'].str.contains('\(', na = False)]
```

```python
censo_demo['populacao'] = censo_demo['populacao'].astype(int)
```

```python
# agrupando por UF e somando a população
censo_demo = censo_demo.groupby('UF')[['populacao']].sum().reset_index()
censo_demo
```

```python
emissao_estados = emissoes_por_ano[emissoes_por_ano['Ano'] == 2021].groupby('Estado')[['Emissão']].sum().reset_index()
emissao_estados
```

```python
# unindo os dois DataFrames
dados_agrupados = pd.merge(emissao_estados, censo_demo, left_on = 'Estado', right_on = 'UF')
dados_agrupados
```

```python
# plotando um scatter para ver se há relação entre população e emissão
dados_agrupados.plot(x = 'populacao', y= 'Emissão', kind = 'scatter', figsize=(8,6));
```

```python
import plotly.express as px
```

```python
# plotando um scatter com PX para ver os estados relativos as bolinhas
px.scatter(data_frame = dados_agrupados, x = 'populacao', y = 'Emissão', text = 'Estado', opacity = 0)
```

```python
# adicionando dado de emissão per capita
dados_agrupados = dados_agrupados.assign(emissao_per_capita = dados_agrupados['Emissão']/dados_agrupados['populacao']).sort_values('emissao_per_capita', ascending = False)
dados_agrupados
```

```python
# plotando um barh para ver a emissao per capita
dados_agrupados.plot(x = 'Estado', y = 'emissao_per_capita', kind = 'barh', figsize=(10,8))
```

```python
px.scatter(data_frame = dados_agrupados, x = 'populacao', y = 'Emissão', text = 'Estado', size = 'emissao_per_capita')
```
